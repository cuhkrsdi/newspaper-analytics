###Define function to get text into dataframe

```python
# Input: a list of file path/url
# out put: a dataframe containing the text extracted from the files, with attributes of filename, newspaper name, publish year, publish month day


def path_to_df(filepath_list):
    counter = 0
    whole_df = pd.DataFrame(columns=["filename", "content", "newspaper", "year", "month_day"])
    whole_df = whole_df.astype(str)
  
    """
    # Original Code for local use, runs on list of paths 
    for path in raw_filepath_list:
    txt_lines = list()
    with open(path, 'r', encoding='utf-8-sig') as file:
        for line in file:
            #print(line.rstrip())
            txt_lines.append(line)
    """

    # modified for URL demo~~~~~~~~~~~~
    for url in filepath_list:
        txt_lines = list()
        file = urlopen(url)
        for line in file:
            decoded_line = line.decode("utf-8")
            txt_lines.append(decoded_line)
            # modified code ends ~~~~~~~~~~~~~~~~~~~

           
        # edit the list of lines into a long string with no \n \r
        join_str = ''.join(txt_lines)
        content = join_str.replace("\r", "").replace("\n", "")

        """
        # Original Code for local use, get filename from filepath
        filename = os.path.splitext(path.name)[0]
        """

        # modified for URL demo~~~~~~~~~~~~~~~
        without_file_extension = os.path.splitext(url)[0]  
        filename = without_file_extension.rsplit('/', 1)[-1]
        # modified code ends ~~~~~~~~~~~~~~~~~~~
        
        newspaper = filename[: filename.find("19")]
        year = filename[filename.find("19") : filename.find("19") + 4]
        month_day = filename[filename.find("19") + 4 : filename.find("19") + 8]
        row = [filename, content, newspaper, year, month_day]
    
        # append list to df as row 
        whole_df.loc[len(whole_df)] = row
        print("--- %s seconds ---" % (time.time() - start_time))
        print(counter)
        counter += 1
    
    return whole_df
```

###Rearrange the result into nested structure

```python
df['filename']
```

```python
# Get year-month-day from file name
ymd_filename_list = [name[name.find("19") : name.find("19") + 8] for name in df['filename']]
```

```python
# take a look whats inside
ymd_filename_list
```

```python
# input: 
# output: ymd_frequency = {ymd: {word: {type1: freq, type2:freq, ...}}}

"""
Since the order of the result is exactly the SAME as order of the original data frame
Hence, it is possible to iterate the newly created list of year-month-day and the result dataframe jointly
"""

# modified for demo 
ymd_frequency_dict = dict()
for rowIndex, row in reset_df.iterrows():
  # set filename from dropped list as variable as key
  ymd_key = ymd_filename_list[rowIndex]
  print(f"Progress: {rowIndex}") # check progress of rows 
  for columnIndex, value in row.items():
    try:
      ner_word = value.word  # value[0]
      ner_type = value.ner  # value[1]
      #
      if ymd_key not in ymd_frequency_dict:
        #
        ymd_frequency_dict[ymd_key] = dict()
      #
      if ner_word not in ymd_frequency_dict[ymd_key]:
        #
        ymd_frequency_dict[ymd_key][ner_word] = dict()
      # 
      if ner_type not in ymd_frequency_dict[ymd_key][ner_word]:
        ymd_frequency_dict[ymd_key][ner_word][ner_type] = ymd_frequency_dict[ymd_key][ner_word].get(ner_type, 1)
      else:
        ymd_frequency_dict[ymd_key][ner_word][ner_type] = ymd_frequency_dict[ymd_key][ner_word].get(ner_type, 1) + 1
    except (TypeError, AttributeError): # skip the empty cells in dataframe
      continue

```

```python
# Structure:    ymd_freq_dict = {yearMonthDay1: {word1: {nerType1: frequency, nerType2:frequency, ...}, word2: {...}}, yearMonthDay2: {...}}
# the scattered data is now reorganized into the new nested structure based on the year-month-day, words, ner type, frequency of occurrence 
ymd_frequency_dict.keys()
```
